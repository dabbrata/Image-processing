# Tapestry with bilinear interpolation

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import math

def bilinear_interpolation(img, x, y):
    height, width, channels = img.shape
    x_floor = int(np.floor(x))
    y_floor = int(np.floor(y))
    x_ceil = min(x_floor + 1, width - 1)
    y_ceil = min(y_floor + 1, height - 1)
    dx = x - x_floor
    dy = y - y_floor

    top_left = img[y_floor, x_floor] * (1 - dx) * (1 - dy)
    top_right = img[y_floor, x_ceil] * dx * (1 - dy)
    bottom_left = img[y_ceil, x_floor] * (1 - dx) * dy
    bottom_right = img[y_ceil, x_ceil] * dx * dy
    
    interpolated_pixel = top_left + top_right + bottom_left + bottom_right
    
    return interpolated_pixel.astype(np.uint8)

def Tapestry(img, a, tx, ty):
    out = np.zeros_like(img)
    img_h, img_w = img.shape[:2]
    out_h, out_w = out.shape[:2]
    xc, yc = out_w // 2, out_h // 2

    for i in range(img_h):
        for j in range(img_w):
            x = j + a * math.sin((2 * np.pi) * (j - xc) / tx)
            y = i + a * math.sin((2 * np.pi) * (i - yc) / ty)
            if 0 <= x < img_w and 0 <= y < img_h:
                out[i][j] = bilinear_interpolation(img, x, y)

    return out

img = cv2.imread(r"C:\Users\WIN\Desktop\image_lab\lab5b\Picture1.png")
tx, ty = 30, 30
a = 5

out = Tapestry(img, a, tx, ty)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 15))
ax1.imshow(img[:, :, ::-1])
ax1.set_title("Input")
ax2.imshow(out[:, :, ::-1])
ax2.set_title("Output")

# cv2.imshow('Biliniear-interpolation',out)
# cv2.imshow('Without bilinear-interpolation',out2)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

plt.show()
```
