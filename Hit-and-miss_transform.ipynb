```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python

A = cv2.imread(r"C:\Users\WIN\Desktop\image_lab\lab5b\input.jpg", 0)
r, A = cv2.threshold(A, 130, 255, cv2.THRESH_BINARY)#_INV)
w = np.ones((150,150),np.uint8)*255
# Thresholding
x1 = np.array(([0,0,0],
               [1,1,0],
               [1,0,0]))

x2 = np.array(([0,1,1],
               [0,0,1],
               [0,0,1]))

x3 = np.array(([1,1,1],
               [0,1,0],
               [0,1,0]))

rate = 50
kernel1 = cv2.resize(x1, None, fx = rate, fy = rate, interpolation = cv2.INTER_NEAREST)
kernel1 = ((kernel1*255).astype(np.uint8))
# r1, kernel1 = cv2.threshold(kernel1, 0, 255, cv2.THRESH_BINARY)

kernel2 = cv2.resize(x2, None, fx = rate, fy = rate, interpolation = cv2.INTER_NEAREST)
kernel2 = ((kernel2*255).astype(np.uint8))
# r2, kernel2 = cv2.threshold(kernel2, 0, 255, cv2.THRESH_BINARY)

kernel3 = cv2.resize(x3, None, fx = rate, fy = rate, interpolation = cv2.INTER_NEAREST)
kernel3 = ((kernel3*255).astype(np.uint8))
# r3, kernel3 = cv2.threshold(kernel3, 0, 255, cv2.THRESH_BINARY)

B1 = kernel1
B2 = w-kernel1
eroded1 = cv2.erode(A,B1,iterations = 1)
eroded2 = cv2.erode((255-A),B2,iterations = 1)


B1 = kernel2
B2 = w-kernel2
eroded1s = cv2.erode(A,B1,iterations = 1)
eroded2s = cv2.erode((255-A),B2,iterations = 1)

B1 = kernel3
B2 = w-kernel3
eroded1t = cv2.erode(A,B1,iterations = 1)
eroded2t = cv2.erode((255-A),B2,iterations = 1)


kernel = np.ones((5,5))
result1 = eroded1&eroded2
dilated1 = cv2.dilate(result1,kernel,iterations = 2)
result2 = eroded1s&eroded2s
dilated2 = cv2.dilate(result2,kernel,iterations = 2)
result3 = eroded1t&eroded2t
dilated3 = cv2.dilate(result3,kernel,iterations = 2)


fig,(ax) = plt.subplots(2,2,figsize=(10,10))
ax[0,0].imshow(A,cmap='gray')
ax[0,0].set_title("Input")
ax[0,1].imshow(kernel1,cmap='gray')
ax[0,1].set_title("kernel1")
ax[1,0].imshow((eroded1&eroded2),cmap='gray')
ax[1,0].set_title("Hit-Miss1")
ax[1,1].imshow((dilated1),cmap='gray')
ax[1,1].set_title("Hit-Miss1-dilated")

```
